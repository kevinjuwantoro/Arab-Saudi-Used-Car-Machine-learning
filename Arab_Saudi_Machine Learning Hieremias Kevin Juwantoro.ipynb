> ## **Metric Evaluation & The Reason of choosing it**
---

**In the context of predicting used car prices there are 3 recommended Metrics that are allign with the objective are RMSE, MAE, and MAPE**

**1. RMSE (Root Mean Squared Error):**
<img src = 'https://arize.com/wp-content/uploads/2023/08/RMSE-equation.png'>

- Description: RMSE measures the average magnitude of errors between predicted and actual values, with an emphasis on larger errors.

- Applicability: In the context of Syarah.com, where predicting used car prices accurately is crucial, RMSE is valuable. Large errors could lead to significant financial losses, making it essential to prioritize and minimize these errors.

**2. MAE (Mean Absolute Error):**
<img src = 'https://miro.medium.com/v2/resize:fit:723/1*9BhnZiaHkApC-gQt3rYpMQ.png'>

- Description: MAE calculates the average absolute difference between predicted and actual values, providing a measure of the average prediction error.

-  Applicability: MAE is useful for evaluating the average accuracy of price predictions on Syarah.com. It considers errors of all magnitudes without emphasizing larger errors, offering a balanced view of model performance.

- Understanding the distribution of prediction errors is crucial. If the errors are normally distributed, metrics like RMSE and MSE might be appropriate. **However, if there are outliers or the distribution is skewed, metrics like MAE might provide a clearer picture of the average error.**

**3. R-Squared (Coefficient of Determination):**

<img src = 'https://4.bp.blogspot.com/-MnLmzZwH2yM/VK87ChxUdII/AAAAAAAAAoc/psiht0wVC4U/w1200-h630-p-k-no-nu/Picture1.png'>

- Description: R-Squared indicates the proportion of variance in the target variable explained by the model. It ranges from 0 to 1, where higher values indicate better explanatory power.

- Applicability: For Syarah.com, R-Squared helps assess how well the model captures the variation in used car prices. A higher R-Squared suggests that the model effectively explains the observed price fluctuations.

**4. MAPE (Mean Absolute Percentage Error):**
<img src = 'https://media.geeksforgeeks.org/wp-content/uploads/20211120204908/mapeformula.png'>

- Description: MAPE calculates the average percentage difference between predicted and actual values, providing insight into the relative error.

- Applicability: In the Syarah.com context, MAPE offers a perspective on the proportional accuracy of price predictions. It helps gauge how well the model performs in predicting prices relative to the actual values.

- This can be beneficial for communication with stakeholders who may not be familiar with the technicalities of machine learning metrics.

**It's often useful to consider multiple metrics to get a comprehensive view of model performance. While RMSE, MAE, and R-Squared provide different perspectives, using a combination of them can offer a more nuanced evaluation.**

**There are others Metric that i think is not recommended to use in the Case for prediciting the right and accurate Price for syarah.com**

**1. MSE (Mean Squared Error):**
<img src ='https://miro.medium.com/v2/resize:fit:1198/1*BtVajQNj29LkVySEWR_4ww.png'>

**- Reason for Not Recommending:** MSE places more emphasis on larger errors due to the squared difference. In the case of Syarah.com, where large prediction errors could have significant consequences, MSE might not be ideal as it could disproportionately influence the overall evaluation.

The reason is that MSE places more emphasis on larger errors because it squares the difference between the prediction and the actual value. In this case, if the model makes a very large prediction error on some examples, the MSE value will be very high, which may not reflect the overall performance of the model.

> ## **IQR (Interquartile Range)**

To determine which data points are considered as outliers, we might use statistical measures like the IQR (Interquartile Range). The IQR is calculated as the difference between the 75th percentile (Q3) and 25th percentile (Q1) values of a dataset. Data points that fall below Q1 - 1.5IQR or above Q3 + 1.5IQR could be considered as outliers.

```python
# Dictionary to store outliers' information
outliers_info = {}

# List of numerical columns
numerical_columns = ['Year', 'Engine_Size', 'Mileage', 'Price']

for column in numerical_columns:
    # Calculate Q1, Q3, and IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Calculate bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers_condition = (df[column] < lower_bound) | (df[column] > upper_bound)
    outliers = df[outliers_condition]

    # Store information in the dictionary
    outliers_info[column] = {
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'number_of_outliers': outliers_condition.sum(),  # Count of outliers
        'outliers_indices': outliers.index.tolist()  # Indices of outliers
    }


# summary for outliers_summary

outliers_summary = {key: {'bounds': (value['lower_bound'], value['upper_bound']), 
                          'number_of_outliers': value['number_of_outliers']} 
                    for key, value in outliers_info.items()}

# Creating DataFrame from 'outliers_summary'
outliers_summary_df = pd.DataFrame.from_dict(outliers_summary, orient='index')

# Splitting 'bounds' into 'lower_bound' and 'upper_bound' columns
outliers_summary_df[['lower_bound', 'upper_bound']] = outliers_summary_df['bounds'].apply(pd.Series)

# Dropping the original 'bounds' column
outliers_summary_df.drop('bounds', axis=1, inplace=True)

# Reordering the columns for a better view
outliers_summary_df = outliers_summary_df[['lower_bound', 'upper_bound', 'number_of_outliers']]

# Display DataFrame as a styled table
styled_table = outliers_summary_df.style.format('{:.0f}')
styled_table

```

> ## **Multicollinearity Test**
---

Multicollinearity is a phenomenon where two or more features in a regression model have a high correlation with each other. This can cause problems in interpreting the results of the model, as it can be difficult to determine the individual contribution of each feature to the target variable. In this context, we need to check if there is a high correlation between other features in the dataset that could affect the interpretation of the model.

To check for multicollinearity, we can use a correlation matrix among the existing features in the dataset. If there is a high correlation between two or more features, it indicates the presence of multicollinearity.

In addition, we can also use the Variance Inflation Factor (VIF) to measure the level of multicollinearity among features. If the VIF of a feature exceeds a certain threshold, it indicates significant multicollinearity with other features in the model.

1. Correlation Matrix: Wev can create a correlation matrix for wer dataset to identify any high correlations between features. If any pair of features has a correlation coefficient close to 1 or -1, it indicates a high degree of multicollinearity.

2. Variance Inflation Factor (VIF): After identifying potential multicollinearity from the correlation matrix, we can calculate the VIF for each feature to quantify the severity of multicollinearity. A common rule of thumb is that if the VIF is greater than 4, the multicollinearity is high.

3. Addressing Multicollinearity: If multicollinearity is found, we might need to address it to improve this model’s performance and interpretability. This could involve removing one of the correlated features, combining correlated features, or using regularization techniques.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python

# untuk menampilkan nilai VIF masing-masing feature
def calc_vif(X):
    vif = pd.DataFrame()
    
    vif['feature'] =  X.columns
    vif['VIF'] =[variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
    vif['acceptance'] = np.where(vif.VIF < 4, 'Yes', 'No')
    
    return vif
```

> ## **A more Brief Explanation for the Chosen Metrix**

**Root Mean Square Error (RMSE):**
- Advantages: It is sensitive to larger errors and can reveal how tightly the observed data clusters around the predicted values1. It’s also easy to interpret since the units of measurement match the target variable.
- Disadvantages: It is sensitive to outliers and can exaggerate results if there are outliers in the dataset.

**Mean Absolute Error (MAE):**
- Advantages: It’s simple to understand and calculate. The result will have the same unit as that of the output.
- Disadvantages: It’s not differentiable at all points, which can be a problem for certain optimization algorithms.

**Mean Absolute Percentage Error (MAPE):**
- Advantages: It’s easy to understand as the error is a percentage. It can be more beneficial because it shows the percentage difference.
- Disadvantages: It’s not possible to use when actual values can be at or close to zero. It’s also asymmetric, meaning that it reports higher errors if the predicted value is higher than the actual value and reports lower errors if the predicted value is lower than the actual.

**Mean Squared Error (MSE):**
- Advantages: It’s differentiable at all points, which is beneficial for optimization algorithms during model training.
- Disadvantages: It’s highly sensitive to outliers, as squaring the errors amplifies their impact on the overall metric.

**Root Mean Squared Percentage Error (RMSPE):**
- Advantages: It measures the ratio of predicted and actual values. It’s preferable when targets have exponential growth, such as population counts, average sales of a commodity over a span of years, etc.
- Disadvantages: Specific disadvantages of RMSPE are not widely discussed in the literature, but it’s important to note that like RMSE, it may also be sensitive to outliers.

**Root Mean Squared Logarithmic Error (RMSLE):**
- Advantages: It’s robust to the effect of outliers and can be beneficial if we do not want to penalize big differences between predicted and actual values when both are very big numbers.
- Disadvantages: The error value is not given in terms of the target, making it difficult to interpret.

**R-Squared (R²):**
- Advantages: It represents the proportion of the variance in the dependent variable which is explained by the model. It’s a scale-free score, meaning irrespective of the values being small or large, the value of R² will be less than one.
- Disadvantages: It can only be used to evaluate linear models19. It also assumes that every independent variable in the model is of equal importance.

It’s a simple and straightforward metric that can give me a basic idea of my model’s performance. MAE measures the average magnitude of the errors in a set of predictions, without considering their direction. It’s the average over the test sample of the absolute differences between prediction and actual observation where all individual differences have equal weight. As a result, MAE is a simple and straightforward metric that provides an idea of how wrong wer predictions were.

**3. Mean Absolute Percentage Error (MAPE):** This metric expresses the forecast error as a percentage, making it easy to interpret and understand. MAPE is a measure that expresses forecast error as a percentage, making it easy to interpret and understand. It expresses the absolute value of the fractional prediction error. This can be very useful if we want to understand the importance of the prediction errors in proportion to the actual values.

As for the use of R², it’s not typically used with gradient boosting models like CatBoost and tec. R² is a measure of how well the model’s predictions fit the actual data, which is **more suitable for linear regression models.**

Regarding **RMSLE and RMSPE, these metrics are indeed more useful when the target variable (in this case, price) has a wide range. If the range of prices in this dataset is relatively small, then these metrics might not be as informative.**

MSE places more emphasis on larger errors due to the squared difference. In the case of Syarah.com, where large prediction errors could have significant consequences, MSE might not be ideal as it could disproportionately influence the overall evaluation. Although MSE can be a good metric in some cases, in the context of predicting used car prices, MSE might not be ideal. **The reason is that MSE places more emphasis on larger errors because it squares the difference between the prediction and the actual value.** In this case, if the model makes a very large prediction error on some examples, the MSE value will be very high, which may not reflect the overall performance of the model.

```python

from sklearn.metrics import mean_absolute_error, mean_squared_error

# Assuming df_y_pred is already defined
df_y_pred = pd.DataFrame({'y_test': y_test, 'y_test_pred': y_pred_cat_tuning})
df_y_pred.reset_index(drop=True, inplace=True)
df_y_pred.sort_values('y_test', inplace=True)

def calculate_metrics(df):
    if len(df) == 0:
        return 0, 0, 0, 0  # Return default values when the subset is empty
    mae = mean_absolute_error(df['y_test'], df['y_test_pred'])
    rmse = np.sqrt(mean_squared_error(df['y_test'], df['y_test_pred']))
    mape = mean_absolute_percentage_error(df['y_test'], df['y_test_pred']) * 100
    length = len(df)
    return mae, rmse, mape, length

ranges = [(5000, 10000), (10000, 20000), (20000, 30000), (30000, 40000),
          (40000, 50000), (50000, 100000), (100000, 150000), (150000, 200000),
          (200000, 250000), (250000, 300000), (300000, 350000), (350000, 400000),
          (400000, 450000), (450000, 500000), (500000, float('inf'))]

list_range2 = ['{}-{}'.format(start, end) for start, end in ranges[:-1]] + ['>{}'.format(ranges[-1][0])]

list_metrics = []

for start, end in ranges:
    subset = df_y_pred[(df_y_pred['y_test'] >= start) & (df_y_pred['y_test'] < end)]
    mae, rmse, mape, length = calculate_metrics(subset)
    list_metrics.append((mae, rmse, mape, length))

columns = ["Score MAE (SAR)", "Score RMSE (SAR)", "Score MAPE (%)", "total Data"]
range_metrics = pd.DataFrame(list_metrics, columns=columns)
range_metrics["Range Price (SAR)"] = list_range2

# Reorder columns for better readability
range_metrics = range_metrics[["Range Price (SAR)", "Score MAE (SAR)", "Score RMSE (SAR)", "Score MAPE (%)", "total Data"]]

range_metrics
```
